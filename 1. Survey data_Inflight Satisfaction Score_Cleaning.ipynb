```python
pd.reset_option('display.max_rows')
common_columns = ['flight_number', 'origin_station_code', 'destination_station_code', 'record_locator', 'scheduled_departure_date']

# Iterate through DataFrame ss12 rows
j=0
for index, row_ss12 in ss12.iterrows():
    # Find the corresponding row in ss1_ based on common columns
    match_indices = ss1_[ss1_[common_columns].eq(row_ss12[common_columns]).all(axis=1)].index.tolist()
    
    for i in match_indices:
        if pd.isnull(ss1_.at[i, 'item_selected']):
            ss1_.at[i, 'item_selected'] = row_ss12['score']

    j=j+1
    if j%500==0:
        print(j)
    
```

```python
ss1_['item_selected'].fillna('Not-Asked', inplace=True)
```
